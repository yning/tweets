```python
import json
import os
import numpy as np
import random
import pandas as pd
import codecs
import re
import pdb
import pickle
import itertools
import nltk
import time
import datetime
import urllib
import numbers
import pytz
from nltk.util import ngrams
from datetime import datetime, timedelta
from collections import OrderedDict
from pandas import Series, DataFrame
def read_tweets_metric_separately(input_metric_path):
	""" Read downloaded metrics from topsy and
		Save sum of hourly count for all news into the dictionary
		Call topsy_download_metrics_for_GSRnews() first
	"""
	gsr_news_profile = dict()
	non_gsr_news_profile = dict()
	all_nodes_files = os.listdir(input_metric_path)
	for node_file in all_nodes_files:
		fname = node_file.strip('.json')
		finfo = fname.split('_')
		gsr_tag = int(finfo[1])
		if gsr_tag == 1 and fname not in gsr_news_profile:
			gsr_news_profile[fname] = dict()
		elif gsr_tag == 1 and fname not in non_gsr_news_profile:
			non_gsr_news_profile[fname] = dict()
		s = []
		with open(input_metric_path + node_file, "rb") as reader:
			js = [json.loads(line.strip()) for line in reader if line.strip()!= ""]
			for j in js:
				try:
					metrics = j['response']['results'][1]['data']
					stats = j['response']['results'][1]['stats']
				except:
					continue
				tempdf = pd.DataFrame(metrics)
				mentions = tempdf.mentions
				s.append(mentions.values)
		df = pd.DataFrame(s)
		if gsr_tag == 1:
			gsr_news_profile[fname] =  {'hourly_count': df.sum(axis=0)}#df.sum(axis=0)
		else:
			non_gsr_news_profile[fname] =  {'hourly_count': df.sum(axis=0)}#df.sum(axis=0)
	return gsr_news_profile, non_gsr_news_profile
```
