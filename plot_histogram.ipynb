```python
from matplotlib import pyplot
import seaborn as sns
sns.set_palette("deep", desat=.6)
sns.set_context(rc={"figure.figsize": (10, 5)})
def plot_histogram_of_mentions_tweets(tweets_profiles, cid):
	for aid, content in tweets_profiles[cid].items():
		profile = content['hourly_count']
		plt.hist(profile, bins=50, alpha = 0.5, label = aid)
		#break
	plt.legend(loc='upper right')
	plt.show()
```
