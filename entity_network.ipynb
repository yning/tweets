```python
def plot_entity_network():

    for cid, nodes in components.items():
        if cid == 512:
            #print "cid1033"
            entity_list = []
            print len(nodes)
            sorted_nodes = OrderedDict(sorted(nodes.items(), key=lambda x:x[1]))
            G = nx.Graph()
            for aid, content in sorted_nodes.items():
                if aid in news2entity:
                    a_entities = news2entity[aid]
                    entity_list.append(a_entities)
                #print entity_list
            mat, entity2id = generate_entity_matrix(entity_list)
            print mat.shape
            alledges = []
            reverse_id2entity = dict()
            for key,value in entity2id.items():
                reverse_id2entity[value] = key
            for i in xrange(mat.shape[0]):
                for j in xrange(mat.shape[1]):
                    if mat[i][j]>0:
                        alledges.append((reverse_id2entity[i], reverse_id2entity[j]))
            #print alledges
            G.add_nodes_from(entity2id.keys())
            G.add_edges_from(alledges)
            nx.draw(G)
            #G.remove_node(entity2id.keys())
            G.remove_edges_from(alledges)
            G.clear()
```
