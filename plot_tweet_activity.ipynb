```python
import json, os, bz2
import numpy as np
import random
import pandas as pd
import codecs
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
#from utils import nstr, public_twittime_to_dt
from datetime import datetime
from collections import OrderedDict
from pylab import *
import re

def read_metric(inputpath):
	c_news_profile = dict()
	all_nodes_files = os.listdir(input_metric_path)
	for node_file in all_nodes_files:
		node_meta = node_file.split("_")
		if len(node_meta) == 6:
			#print node_file
			component_id = node_meta[1]
			article_id = node_meta[3] + "_" + node_meta[4]+ "_" + node_meta[5]
			if component_id not in c_news_profile:
				c_news_profile[component_id] = dict()
				c_news_profile[component_id][article_id] = []
			else:
				c_news_profile[component_id][article_id] = []
			s = []
			with open(input_metric_path + node_file, "rb") as reader:
				js = [json.loads(line.strip()) for line in reader if line.strip()!= ""]
				for j in js:
					metrics = j['response']['results'][1]['data']
					stats = j['response']['results'][1]['stats']
					peakepoch = stats['peak']['mentions']
					tempdf = pd.DataFrame(metrics)
					mentions = tempdf.mentions
					#peaktime = datetime.fromtimestamp(peakepoch)
					#print mentions.values
					s.append(mentions.values)
			df = pd.DataFrame(s)
			#print df.shape
			c_news_profile[component_id][article_id] = df.sum(axis=0)
			#print c_news_profile[component_id][article_id].index
			#break
	return c_news_profile
```
