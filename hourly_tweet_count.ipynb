```python
import json, os, bz2
import numpy as np
import random
import pandas as pd
import codecs
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
#from utils import nstr, public_twittime_to_dt
from datetime import datetime
from collections import OrderedDict
import re

def plot_hourly_tweet_count(input_path):
	months = mdates.MonthLocator()
	days =  mdates.DayLocator()
	hours = mdates.HourLocator()
	daysFmt = mdates.DateFormatter('%d')
	monthsFmt = mdates.DateFormatter('%m')
	hoursFmt = mdates.DateFormatter('%H')

	PUBLIC_TWITTER_TIME_FORMAT = "%a %b %d %H:%M:%S +0000 %Y"

	allfiles = os.listdir(input_path)
	counter = 0
	fig = plt.figure(figsize(15, 6), dpi=80)
	ax1 = fig.add_subplot(111)
	ax1.set_position([0.1,0.1,0.6,0.8])
	colormap = plt.cm.gist_ncar
	plt.gca().set_color_cycle([colormap(i) for i in np.linspace(0, 0.9, 20)])
	labels = []
	for input_file in allfiles:
		#if counter > 5:
		#	break
		#else:
		#	counter += 1
		time2count = dict()
		with open(input_path + input_file, "rb") as jsonreader:
			for line in jsonreader:
				j = json.loads(line.strip())
				if 'created_at' in j:
					d = datetime.strptime(j['created_at'], PUBLIC_TWITTER_TIME_FORMAT)
					if d not in time2count:
						time2count[d] = 1
					else:
						time2count[d] += 1


		#ts = Series(time2count.values(), time2count.keys())
		df = DataFrame(data=time2count.values(), index=time2count.keys())
		grouped = df.groupby( [df.index.year, df.index.month, df.index.day, df.index.hour] )#.count()
		#print len(grouped)

		x = []
		relative_x = [i for i in range(-len(grouped)/2, len(grouped)/2)]
		#print len(relative_x)
		y = []
		hours2count = dict()
		for hour, count in grouped:
			#print hour[3]
			x.append(datetime.strptime("-".join([str(hour[i]) for i in xrange(len(hour) - 1)]) + "T" + str(hour[3]),'%Y-%m-%dT%H'))
			#print count.sum()
			y.append(count.sum())


		#ax1.xaxis.set_major_locator(hours)
		#ax1.xaxis.set_major_formatter(hoursFmt)
		plt.xlabel('Time(-24:24)')
		plt.ylabel('Hourly Tweet Count')
		embersid = input_file.split("_")[4] + "_" + input_file.split("_")[5].strip(".json") 
		labels.append(embersid)
		plt.plot(relative_x, y, linewidth = 3)

	plt.legend(labels, ncol=2, loc='upper center', bbox_to_anchor=[0.5, 1.1], columnspacing=1.0, labelspacing=0.0,\
		handletextpad=0.0, handlelength=1.5,\
		fancybox=True, shadow=True)
	leg = ax1.legend()
	fig.autofmt_xdate()
	plt.show()
component_list = ["/home/yning/story-twitter/topsy/component1033/",\
"/home/yning/story-twitter/topsy/component1031/",\
"/home/yning/story-twitter/topsy/component10/"]
for inputpath in component_list:
	plot_hourly_tweet_count(inputpath)
```
