```python
import json
import os
import numpy as np
import random
import pandas as pd
import codecs
import re
import pdb
import pickle
import itertools
import nltk
import time
import datetime
import urllib
import numbers
import pytz
from nltk.util import ngrams
from datetime import datetime, timedelta
from collections import OrderedDict
from pandas import Series, DataFrame
def read_gsr_metric(input_metric_path):
	""" Read downloaded metrics from topsy and
		Save sum of hourly count for all news into the dictionary
		Call topsy_download_metrics_for_GSRnews() first
	"""
	news_profile = dict()
	all_nodes_files = os.listdir(input_metric_path)
	for node_file in all_nodes_files:
		article_id = node_file.strip('.json')
		if article_id not in news_profile:
			news_profile[article_id] = dict()
		s = []
		with open(input_metric_path + node_file, "rb") as reader:
			js = [json.loads(line.strip()) for line in reader if line.strip()!= ""]
			for j in js:
				try:
					metrics = j['response']['results'][1]['data']
					stats = j['response']['results'][1]['stats']
				except:
					continue
				#peakepoch = stats['peak']['mentions']
				tempdf = pd.DataFrame(metrics)
				mentions = tempdf.mentions
				#peaktime = datetime.fromtimestamp(peakepoch)
				#print mentions.values
				s.append(mentions.values)
		df = pd.DataFrame(s)
		#print df.shape
		news_profile[article_id] =  {'hourly_count': df.sum(axis=0)}#df.sum(axis=0)
		#print c_news_profile[component_id][article_id].index
		
	return news_profile
```
